```python
def deserialize(serialized_data: bytes) -> list[float]:
    """deserializes raw bytes back into a list of floats"""
    num_floats = len(serialized_data) // struct.calcsize('f')  # number of floats
    return list(struct.unpack(f'{num_floats}f', serialized_data))
```
